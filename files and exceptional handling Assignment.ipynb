```python
Q.1    How can you open a file for writing in Python and write a string to it
ANS:   with open("example.txt", "w") as file:
    file.write("Hello, world!")



Q.2    Write a Python program to read the contents of a file and print each line
ANS:   with open("example.txt", "r") as file:
    for line in file:
        print(line.strip())


Q.3     How would you handle a case where the file doesnt exist while trying to open it for reading
ANS:    try:
    with open("example.txt", "r") as file:
        content = file.read()
        print(content)
except FileNotFoundError:
    print("Error: The file does not exist.")




Q.4     Write a Python script that reads from one file and writes its content to another file
ANS:    with open("source.txt", "r") as source_file, open("destination.txt", "w") as dest_file:
    dest_file.write(source_file.read())




Q.5     How would you catch and handle division by zero error in Python
ANS:    try:
    result = numerator / denominator
except ZeroDivisionError:
    result = float('inf')  # or any other default value or action
    print("Cannot divide by zero!")




Q.6    Write a Python program that logs an error message to a log file when a division by zero exception occurs
ANS:   import logging

# Configure logging to write to a file
logging.basicConfig(filename='error_log.txt', level=logging.ERROR,
                    format='%(asctime)s - %(levelname)s - %(message)s')

def divide_numbers(a, b):
    try:
        result = a / b
        return result
    except ZeroDivisionError as e:
        logging.error("Attempted to divide by zero: %s", e)
        return None

# Example usage
numerator = 10
denominator = 0
result = divide_numbers(numerator, denominator)

if result is None:
    print("Division by zero occurred. Check the log file for details.")
else:
    print(f"The result is: {result}")





Q.7    How do you log information at different levels (INFO, ERROR, WARNING) in Python using the logging module
ANS:   import logging

# Configure logging to display messages at the INFO level and above
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

logging.debug("This is a debug message")   # Won't be shown (below INFO level)
logging.info("This is an info message")    # Will be shown
logging.warning("This is a warning message") # Will be shown
logging.error("This is an error message")  # Will be shown
logging.critical("This is a critical message") # Will be shown




Q.8     Write a program to handle a file opening error using exception handling
ANS:     import os

def read_file(path):
    try:
        with open(path, "r") as file:
            content = file.read()
    except FileNotFoundError:
        print(f"Error: File '{path}' not found.")
    except PermissionError:
        print(f"Error: Insufficient permissions to read '{path}'.")
    except IsADirectoryError:
        print(f"Error: '{path}' is a directory, not a file.")
    except OSError as e:
        print(f"An OS error occurred while opening '{path}': {e}")
    else:
        print("File read successfully. Content:")
        print(content)
    finally:
        print("Finished attempting to read the file.")

if __name__ == "__main__":
    file_path = "example.txt"
    read_file(file_path)





Q.9    How can you read a file line by line and store its content in a list in Python
ANS:   with open('example.txt', 'r') as file:
    lines = file.readlines()  # ['Line 1\n', 'Line 2\n', ...]




Q.10     How can you append data to an existing file in Python
ANS:     with open("example.txt", "a") as f:
    f.write("This line is appended.\n")




Q.11    Write a Python program that uses a try-except block to handle an error when attempting to access a
dictionary key that doesn't exist
ANS:   def get_value(d: dict, key):
    try:
        # Attempt to access the value for the given key
        return d[key]
    except KeyError:
        # Handle the missing key by logging or returning a default message/value
        print(f"Error: Key '{key}' not found in the dictionary.")
        return None

# Example usage:
person = {
    "name": "Alice",
    "age": 30,
    "city": "Paris"
}

print(get_value(person, "age"))      # Outputs: 30
print(get_value(person, "country"))  # Outputs: Error... then None




Q.12    Write a program that demonstrates using multiple except blocks to handle different types of exceptions
ANS:    def process_user_input():
    try:
        num = int(input("Enter an integer divisor: "))
        result = 10 / num
    except ZeroDivisionError:
        print("Error: You can't divide by zero!")
    except ValueError:
        print("Error: Invalid input – please enter a valid integer.")
    except Exception as e:
        # A catch-all fallback for unexpected exceptions
        print(f"An unexpected error occurred: {type(e).__name__} – {e}")
    else:
        print(f"Success: 10 divided by {num} is {result}")
    finally:
        print("Execution of process_user_input is complete.")

if __name__ == "__main__":
    process_user_input()




Q.13      How would you check if a file exists before attempting to read it in Python
ANS:      import os

file_path = "example.txt"

if os.path.exists(file_path):  # True if file or directory exists
    print(f"File '{file_path}' exists; proceeding to read.")
    with open(file_path, 'r') as f:
        data = f.read()
else:
    print(f"Error: File '{file_path}' does not exist.")






Q.14     Write a program that uses the logging module to log both informational and error messages
ANS:     import logging

# Configure logging: INFO and above to console, DEBUG and above to file
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s',
    datefmt='%Y-%m-%d %H:%M:%S',
    handlers=[
        logging.StreamHandler(),              # INFO+ to console
        logging.FileHandler('app.log'),       # INFO+ to app.log
    ]
)

logger = logging.getLogger(__name__)  # Use module name for context

def divide(a, b):
    try:
        result = a / b
    except ZeroDivisionError as e:
        # Logs an ERROR with full stack trace
        logger.error("Division error: attempted %s / %s", a, b, exc_info=True)
        return None
    else:
        # Logs an INFO message about successful operation
        logger.info("Division successful: %s / %s = %s", a, b, result)
        return result

if __name__ == "__main__":
    divide(10, 2)  # INFO logged
    divide(5, 0)   # ERROR logged (with traceback)





Q.15    Write a Python program that prints the content of a file and handles the case when the file is empty
ANS:    import os

def print_file_content(path):
    try:
        # Ensure the file exists before reading
        if not os.path.exists(path):
            print(f"Error: File '{path}' does not exist.")
            return

        # Check if the file is empty
        if os.path.getsize(path) == 0:
            print(f"Notice: File '{path}' is empty.")
            return

        # If it exists and isn't empty, read and print line by line
        with open(path, 'r', encoding='utf-8') as f:
            for line in f:
                print(line, end='')

    except OSError as e:
        # Catch any other file-related errors
        print(f"Error accessing file '{path}': {e}")

if __name__ == "__main__":
    filepath = "example.txt"
    print_file_content(filepath)





Q.16    Demonstrate how to use memory profiling to check the memory usage of a small program
ANS:    Step 1: Install memory_profiler
 pip install memory-profiler

Step 2: Create a small script to profile
Save this as mem_test.py:

from memory_profiler import profile

@profile
def allocate_lists():
    a = [i for i in range(100_000)]        # allocate list of 100k ints
    b = [i * 2 for i in range(200_000)]    # allocate list of 200k ints
    del b                                  # free the second list
    return sum(a)

if __name__ == "__main__":
    allocate_lists()
Use the @profile decorator on any function you want to inspect.


 Step 3: Run the profiler

python -m memory_profiler mem_test.py
You'll receive output like:


Line #    Mem usage    Increment   Occurrences   Line Contents
=============================================================
     3     30.3 MiB     30.3 MiB           1   @profile
     4     32.0 MiB      1.7 MiB           1   a = [i for i in range(100000)]
     5     48.0 MiB     16.0 MiB           1   b = [i * 2 for i in range(200000)]
     6     32.0 MiB    -16.0 MiB           1   del b
     7     32.0 MiB      0.0 MiB           1   return sum(a)






Q.17     Write a Python program to create and write a list of numbers to a file, one number per line
ANS:     def write_numbers_to_file(numbers, filename):
    """
    Writes each number from the list `numbers` to `filename`, one per line.
    """
    with open(filename, "w") as f:
        for num in numbers:
            f.write(f"{num}\n")

if __name__ == "__main__":
    nums = list(range(1, 11))  # Example list: numbers 1 through 10
    write_numbers_to_file(nums, "numbers.txt")
    print("Numbers successfully written to numbers.txt")





Q.18    How would you implement a basic logging setup that logs to a file with rotation after 1MB
ANS:    import logging
from logging.handlers import RotatingFileHandler

# Set up the logger
logger = logging.getLogger('my_logger')
logger.setLevel(logging.DEBUG)

# Create a rotating file handler that rotates at 1MB
handler = RotatingFileHandler('my_log.log', maxBytes=1 * 1024 * 1024, backupCount=3)
handler.setLevel(logging.DEBUG)

# Create a formatter and set it for the handler
formatter = logging.Formatter('%(asctime)s - %(name)s - %(levelname)s - %(message)s')
handler.setFormatter(formatter)

# Add the handler to the logger
logger.addHandler(handler)

# Example log messages
logger.debug('This is a debug message')
logger.info('This is an info message')
logger.warning('This is a warning message')
logger.error('This is an error message')
logger.critical('This is a critical message')




Q.19      Write a program that handles both IndexError and KeyError using a try-except block
ANS:      def access_elements():
    my_list = [10, 20, 30]
    my_dict = {"name": "Alice", "age": 25}

    try:
        # Attempting to access an index that might be out of range
        print(my_list[5])  # This will raise IndexError

        # Attempting to access a key that might not exist
        print(my_dict["address"])  # This will raise KeyError

    except IndexError:
        print("Caught an IndexError: Index is out of range.")

    except KeyError:
        print("Caught a KeyError: Key does not exist in the dictionary.")

if __name__ == "__main__":
    access_elements()
Output:

Caught an IndexError: Index is out of range.
Caught a KeyError: Key does not exist in the dictionary.





Q.20     How would you open a file and read its contents using a context manager in Python
ANS:     with open("example.txt", "r") as file:
    content = file.read()
    print(content)



Q.21      Write a Python program that reads a file and prints the number of occurrences of a specific word
ANS:      def count_word_in_file(filename, target_word):
    try:
        with open(filename, 'r') as file:
            content = file.read().lower()  # Read and convert to lowercase
            word_count = content.split().count(target_word.lower())  # Count occurrences
        return word_count
    except FileNotFoundError:
        print(f"Error: The file '{filename}' was not found.")
        return 0

if __name__ == "__main__":
    filename = "example.txt"  # Replace with your file path
    target_word = "python"    # Replace with the word you want to count
    count = count_word_in_file(filename, target_word)
    print(f"The word '{target_word}' appears {count} times in the file '{filename}'.")




Q.22      How can you check if a file is empty before attempting to read its contents
ANS:      import os

file_path = 'example.txt'

if os.path.exists(file_path) and os.path.getsize(file_path) == 0:
    print("The file is empty.")
else:
    print("The file is not empty.")





Q
```
